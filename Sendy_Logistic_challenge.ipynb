```python
#Calculate harvsine distance from the given coordinates
#Harvsine distance calculates the shortest distance between two point on a sphere
def harvsine_array(lat1, lng1 , lat2, lng2):
    lat1, lng1, lat2, lng2 = map(np.radians, (lat1, lng1 , lat2, lng2))
    AVG_EARTH_RADIUS = 6371 #in kilometers
    lat = lat2 - lat1
    lng = lng2 - lng1
    d = np.sin(lat * 0.5) ** 2 + np.cos(lat1) * np.cos(lat2) * np.sin(lng * 0.5) ** 2
    h = 2 * AVG_EARTH_RADIUS * np.arctan(np.sqrt(d))
    return h
```

```python
train_merge['harvsine_distance'] = harvsine_array(train_merge['Pickup Lat'],train_merge['Pickup Long']
                                                  ,train_merge['Destination Lat'], train_merge['Destination Long'])


test_merge['harvsine_distance'] = harvsine_array(test_merge['Pickup Lat'],test_merge['Pickup Long']
                                                  ,test_merge['Destination Lat'], test_merge['Destination Long'])
```

```python
#Calculate manhattan distance of the coodinates
# Manhattan distance is the sum of the horizontal and vertical distance between points on a grid
def manhattan_dist(lat1 ,lng1, lat2, lng2):
    a = harvsine_array(lat1, lng1 , lat1 , lng2)
    b = harvsine_array(lat1,lng1, lat2, lng1)
    
    return a + b
```

```python
train_merge['Manhattan_distance'] = manhattan_dist(train_merge['Pickup Lat'],train_merge['Pickup Long']
                                                  ,train_merge['Destination Lat'], train_merge['Destination Long'])

test_merge['Manhattan_distance'] = manhattan_dist(test_merge['Pickup Lat'],test_merge['Pickup Long']
                                                  ,test_merge['Destination Lat'], test_merge['Destination Long'])
```

```python
#Direction from the given coordinates
def bearing_array(lat1, lng1, lat2, lng2):
    AVG_EARTH_RADIUS = 6371  # in km
    lng_delta_rad = np.radians(lng2 - lng1)
    lat1, lng1, lat2, lng2 = map(np.radians, (lat1, lng1, lat2, lng2))
    y = np.sin(lng_delta_rad) * np.cos(lat2)
    x = np.cos(lat1) * np.sin(lat2) - np.sin(lat1) * np.cos(lat2) * np.cos(lng_delta_rad)
    return np.degrees(np.arctan2(y, x))

train_merge['Direction'] = bearing_array(train_merge['Pickup Lat'],train_merge['Pickup Long']
                                                  ,train_merge['Destination Lat'], train_merge['Destination Long'])

test_merge['Direction'] = bearing_array(test_merge['Pickup Lat'],test_merge['Pickup Long']
                                                  ,test_merge['Destination Lat'], test_merge['Destination Long'])
```
